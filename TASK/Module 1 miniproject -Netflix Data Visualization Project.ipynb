Which type of content is more available on Netflix?



```python

content_count = df["type"].value_counts()
print(content_count)
```

```python
print("Movie Type Content is More Avaible on Netflix . ")
```

```python
plt.bar(content_count.index, content_count.values)
plt.title("Content Type on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Number of Titles")
plt.show()
```

Which countries contribute the most content to Netflix?


```python
country_count = {}
for countries in df["country"].dropna():
    for country in countries.split(","):
        country = country.strip()

        if country in country_count:
            country_count[country] += 1
        else:
            country_count[country] = 1
```

```python
# Converting to Series and sorting in descending
country_count = pd.Series(country_count).sort_values(ascending=False)
print(country_count.head(10))
```

```python
print("United States contributed most content to netflix that is 3690")
```

```python
plt.bar(country_count.head(5).index, country_count.head(5).values)
plt.title("Top 10 Countries Contributing Content to Netflix")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

Which directors have the most titles on Netflix?


```python
director_count = df["director"].value_counts()
print(director_count)
```

```python
plt.figure(figsize=(10,6))
plt.bar(director_count.head(5).index, director_count.head(5).values)
plt.title("Top  Directors on Netflix")
plt.xlabel("Director")
plt.ylabel("Number of Titles")
plt.show()
```

```python
print(" ------ PROJECT RESULT-------")
print("The most available content type is", content_count.idxmax(),"with", content_count.max(), "titles.")
print("The country contributing the most titles is", country_count.idxmax(),"with", country_count.max(), "titles.")
print("The director with the most titles is", director_count.idxmax(),"with", director_count.max(), "titles.")
```
