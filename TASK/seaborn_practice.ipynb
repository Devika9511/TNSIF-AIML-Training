Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive, informative, and statistical graphics with less code.

sns.scatterplot() – Scatter plot

sns.lineplot() – Line plot

sns.barplot() – Bar plot (average values)

sns.countplot() – Frequency count plot

sns.boxplot() – Box plot

sns.violinplot() – Violin plot

sns.heatmap() – Heatmap

sns.pairplot() – Pairwise relationship plot

sns.histplot() – Histogram / Distribution plot

sns.regplot() – Regression plot

sns.catplot() – Categorical plot

```python
#Importing LIbraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
print("Dataset Information:")
print(iris.info())

print("Statistical Summary:")
print(iris.describe())
```

```python
#Scatter Plot
sns.scatterplot(x="petal_length",y="petal_width",hue="species",data=iris)
plt.title("Sepal length vs sepal width")
plt.show()
```

```python
#Line Plot
sns.lineplot(x="species", y="petal_length", data=iris)
plt.title("Line Plot -species vs petal_length")
plt.show()
```

```python
#Bar Plot
sns.barplot( x="species",y="petal_length",data=iris
)
plt.title("Bar Plot - Average Petal Length by Species")
plt.show()
```

```python
#Count Plot
plt.figure(figsize=(6,4))
sns.countplot(x="species",data=iris)
plt.title("Count Plot - Number of Flowers per Species")
plt.show()
```

```python
#Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x="species",y="petal_length",data=iris)
plt.title("Box Plot - Petal Length Distribution")
plt.show()
```

```python
#Violin Plot
plt.figure(figsize=(6,4))
sns.violinplot(x="species",y="petal_length",data=iris
)
plt.title("Violin Plot - Petal Length Distribution")
plt.show()
```

```python
#Heatmap (Correlation Matrix)
plt.figure(figsize=(6,4))
corr = iris.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm",linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Pair Plot
sns.pairplot(iris,hue="species")
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

```python
#Distribution Plot (Histogram + KDE)
plt.figure(figsize=(6,4))
sns.histplot(iris["petal_length"],kde=True,bins=20)
plt.title("Distribution of Petal Length")
plt.show()
```

```python
#Regression Plot
plt.figure(figsize=(6,4))
sns.regplot(x="petal_length",y="petal_width",data=iris)
plt.title("Regression Plot - Petal Length vs Petal Width")
plt.show()
```
