```python
#Finding Missing Values
df.isnull().sum()
```

```python
#Filling Missing Values
num_cols = ["rainfall","temperature","humidity","wind_speed"]
df[num_cols] = df[num_cols].fillna(df[num_cols].mean()) # for numerical data we are filling with mean
```

```python
#Filling Missing Categorical Values
df["weather_condition"] = df["weather_condition"].fillna(df["weather_condition"].mode()[0]) #for categorical data we use mode
```

```python
#Converting Date Column
df["date"] = pd.to_datetime(df["date"], errors="coerce")
```

```python
#Rename Columns to Lowercase to handle inconsistency
df.columns = df.columns.str.lower()
df["weather_condition"] = df["weather_condition"].str.lower()
```

```python
#Removing Duplicate Rows
df = df.drop_duplicates()
```

```python
#Increase Rainfall by 10%
df["rainfall"] = df["rainfall"] * 1.10
```

```python
df["rain_category"] = df["rainfall"].apply(lambda x: "Heavy Rain" if x > 10 else "Light Rain")#Creates a new feature based on rainfall.
df[df["rainfall"] > 10] #Filtering Dataset , Returns only rows where rainfall is greater than 10 mm.
```

```python
#Droping Unnecessary Column
df = df.drop(columns=["wind_speed"])
```

```python
#One-Hot Encoding-converting text to numerical
df = pd.get_dummies(df, columns=["weather_condition"])
```

```python
#Sorting Dataset By date
df.sort_values(by="date")
```

```python
#Detecing & Handling Outliers
Q1 = df["rainfall"].quantile(0.25)
Q3 = df["rainfall"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df = df[(df["rainfall"] >= lower) & (df["rainfall"] <= upper)]
```
