```python
import pandas as pd
import numpy as np

# Create Dataset
data = {
    "Book_ID": [101, 102, 103, 104, 105,106],
    "Book_Name": ["Python Basics", "Data Science", "Machine Learning",
                  "Deep Learning", "History of India","cs"],
    "Category": ["Programming", "Programming", "AI", "AI", "History","science"],
    "Borrow_Count": [120, 180, np.nan, 160, 90,110],
    "Rating": [4.5, np.nan, 4.9, 4.7, 4.2,np.nan]
}

df = pd.DataFrame(data)

# Inspect Dataset
print("Dataset:")
print(df)

# Handle Missing Values
print(df.isnull().sum())

# Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
# Find Books Borrowed More Than Average
avg_borrow = df["Borrow_Count"].mean()
above_avg = df[df["Borrow_Count"] > avg_borrow]

print("Books Borrowed More Than Average:")
print(above_avg[["Book_Name", "Borrow_Count"]])
```

```python
# Sort by Borrow Count
sorted_books = df.sort_values(by="Borrow_Count", ascending=True)
print(" by Borrow Count:")
print(sorted_books[["Book_Name", "Borrow_Count"]])
```

```python
#Category-wise Average Ratings
print("Category-wise Average Ratings:")
print(df.groupby("Category")["Rating"].mean())
```

```python
# Create Popularity Index
df["Popularity_Index"] = (df["Borrow_Count"] * df["Rating"])/100
print("Popularity Index:")
print(df[["Book_Name", "Popularity_Index"]])
```

```python
# Apply NumPy Functions
print("NumPy Analysis:")
print("Maximum Borrow Count:", np.max(df["Borrow_Count"]))
print("Minimum Borrow Count:", np.min(df["Borrow_Count"]))
print("Average Rating:", np.mean(df["Rating"]))
print("Standard Deviation of Ratings:", np.std(df["Rating"]))
```

```python
# Identify Most Borrowed Book
most_borrowed = df.loc[df["Borrow_Count"].idxmax()]
print("Most Borrowed Book:")
print(most_borrowed[["Book_Name", "Borrow_Count"]])
```

```python
# Generate Report
print("LIBRARY REPORT")
print("Total Books:", len(df))
print("Average Borrow Count:", avg_borrow)
print("Most Borrowed Book:", most_borrowed["Book_Name"])
```
