```python
# create 1d arr and find sum , mean , max,min
arr5 = np.array([10, 20, 30, 40, 50],)
print(arr5)
print("sum =", np.sum(arr5))
print("mean =", np.mean(arr5))
print("Max =", np.max(arr5))
print("Min =", np.min(arr5))
print("std=",np.std(arr5))
```

```python
import numpy as np

# Create Array
arr = np.array([10, 20, 30, 40, 50])
print("Array:", arr)

# Range of Values
a = np.arange(5, 26, 5)
print("Arange:", a)

# Zeros Array
zero = np.zeros((3, 4))
print("Zeros Array:\n", zero)

# Ones Array
one = np.ones((2, 5))
print("Ones Array:\n", one)

# Identity Matrix
identity = np.eye(3)
print("Identity Matrix:\n", identity)
```

```python
# Evenly Spaced Values
line = np.linspace(0, 100, 11)
print("Linspace:", line)

# Random Numbers
rand = np.random.rand(3, 3)
print("Random Array:\n", rand)

# Shape
print("Shape:", rand.shape)

# Reshape
arr2 = np.arange(1, 13).reshape(3, 4)
print("Reshaped Array:\n", arr2)

# Aggregation Functions
print("Sum:", np.sum(arr2))
print("Mean:", np.mean(arr2))
print("Std:", np.std(arr2))

# Min, Max and Indices
print("Min:", np.min(arr2))
print("Max:", np.max(arr2))
print("Argmin:", np.argmin(arr2))
print("Argmax:", np.argmax(arr2))

# Matrix Multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Dot Product:\n", np.dot(A, B))
print("Matrix Multiplication:\n", np.matmul(A, B))

# Transpose
print("Transpose of A:\n", A.T)

# Sorting
nums = np.array([12, 5, 18, 3, 9])
print("Sorted Array:", np.sort(nums))
```

```python
# Importing NumPy
import numpy as np

# Creating Arrays
arr1 = np.array([10, 20, 30, 40, 50])
arr2 = np.array([[2, 4, 6],
                 [8, 10, 12],
                 [14, 16, 18]])

print("1D Array:", arr1)
print("2D Array:\n", arr2)

# Array with specific values
zeros = np.zeros((2, 5))
ones = np.ones((4, 2))
identity = np.eye(4)

print("Zeros:\n", zeros)
print("Ones:\n", ones)
print("Identity Matrix:\n", identity)
```

```python
# Range and linspace
arange_arr = np.arange(5, 30, 3)
linspace_arr = np.linspace(0, 100, 6)

print("Arange:", arange_arr)
print("Linspace:", linspace_arr)

# Random numbers
rand_arr = np.random.rand(3, 3)
print("Random Array:\n", rand_arr)

# Random integers (OTP-like)
otp = ((np.random.rand(5) * 900000) + 100000).astype(int)
print("OTP Numbers:", otp)
```
